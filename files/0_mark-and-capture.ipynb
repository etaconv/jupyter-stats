```python
import matplotlib.pyplot as plt
import numpy as np
import random as random
```

```python
x = 364 # População real de elementos

t = 100 # Nº de tentativas
```

Definição de constantes que vão ser utilizadas mais tarde no programa

```python
def contar_macacos(m: int) -> float:    
    c1, c2 = [set(np.random.choice(m, size=np.random.randint(1, m+1), replace=False) + 1) for _ in range(2)]
    comum = c1 & c2

    return (len(c1) * len(c2)) / len(comum) if comum else 0
```

Definição de uma função que realiza duas captura e escolhe os "macacos" comuns às duas

```python
estimativas = [result for _ in range(t) if (result := contar_macacos(x)) != 0]

medianas = [np.median(estimativas[:i+1]) for i in range(len(estimativas))]
```

execução dessa função `t` vezes, e a realização da mediana, progressivamente, de cada uma das execuções. 

```python
plt.figure(figsize=(10,4))
plt.style.use("_mpl-gallery")
plt.xlabel("Nº de tentativas")
plt.ylabel("Mediana dos resultados")
plt.title("Aplicação do método de captura-recaptura")
plt.plot(medianas)
plt.show()
print(f"Amplitude: {np.max(estimativas) - np.min(estimativas)}")
```

Demonstração dos resultados num gráfico de linhas

(Este gráfico tem uma peculiariedade. Começa no zero. Isto porque listas em Python têm o seu primeiro elemento indexado pelo zero)
