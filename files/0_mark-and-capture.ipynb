```python
import matplotlib.pyplot as plt
import numpy as np
import random as random
```

```python
x = 563 # População real de elementos

t = 130 # Nº de tentativas
```

Definição de constantes que vão ser utilizadas mais tarde no programa

```python
def contar_macacos(m: int) -> float:
    population=range(1,m+1)
    captura = random.sample(population=population, k=random.randint(1, m))
    segunda_captura = random.sample(population=population, k=random.randint(1, m))
    comum = [value for value in captura if value in segunda_captura]

    return 0 if len(comum) == 0 else (len(captura) * len(segunda_captura)) / len(comum)
```

Definição de uma função que realiza duas captura e escolhe os "macacos" comuns às duas

```python
estimativas = [i for i in [contar_macacos(x) for _ in range(t)] if i != 0]

medianas = [np.median(estimativas[:i+1]) for i in range(len(estimativas))]
```

execução dessa função `t` vezes, e a realização da mediana, progressivamente, de cada uma das execuções. 

```python
plt.figure(figsize=(10,4))
plt.style.use("_mpl-gallery")
plt.xlabel("Nº de tentativas")
plt.ylabel("Mediana dos resultados")
plt.title("Aplicação do método de captura-recaptura")
plt.plot(medianas)

plt.show()
print(f"Amplitude: {np.max(estimativas) - np.min(estimativas)}")
```
